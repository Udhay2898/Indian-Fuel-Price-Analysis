```python
# Group by 'Year' and 'Month', and aggregate the sum of petrol and diesel prices for each city
year_gr = df.groupby(['Year', 'Month']).agg({
    'Delhi (P)': 'mean',
    'Mumbai (P)': 'mean',
    'Chennai (P)': 'mean',
    'Kolkata (P)': 'mean',
    'Delhi (D)': 'mean',
    'Mumbai (D)': 'mean',
    'Chennai (D)': 'mean',
    'Kolkata (D)': 'mean'
}).reset_index()

year_gr.head(2)
```

```python

year_gr_pivot_pet = year_gr.pivot(index = 'Year',columns = 'Month', values = ['Delhi (P)'])

year_gr_pivot_dsl = year_gr.pivot(index = 'Year',columns = 'Month', values =  ['Delhi (D)',	'Mumbai (D)',	'Chennai (D)', 'Kolkata (D)'])
```

```python
year_gr_pivot_pet.plot.line(figsize = (10,5), marker = '*',grid = True)
```
